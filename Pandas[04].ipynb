```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris_datasets = sns.load_dataset('iris')
iris_datasets.head()
```

```python
iris_datasets.info()
```

```python
# 각 종별로 가장 큰 값과 가장 작은 값의 비율 구하기
# agg() : 함수를 인자로 받아서 함수실행
# -> 여러 개의 함수에 대한 결과 출력 시 리스트 형태로 담아서 결과 확인가능
def get_ratio(x) :
    return x.max() / x.min()
```

```python
grp_iris_species = iris_datasets.groupby(iris_datasets['species'])
```

```python
for key, group in grp_iris_species :
    print('key : ',key)
    print('number : ',len(group))
    print(group)
    print('\n')
```

```python
grp_iris_species.agg(get_ratio)
```

```python
# 각 종별 요약통계량 출력
grp_iris_species.describe().T
```

```python
# 종별로 가장 긴 꽃잎길이(petal_length)가 큰 3개의 데이터만 출력
# sort_values() -> 내림차순 정렬
# apply() : 그룹연산이 아닌 경우 사용
# agg() : 그룹연산인 경우 사용 (통계연산)
# display(iris_datasets)
iris_datasets.sort_values(by='petal_length',ascending=False)[:3]
```

```python
def max3_petal_length(df) :
    return df.sort_values(by='petal_length',ascending=False)[:3]
```

```python
iris_datasets.groupby(iris_datasets.species).apply(max3_petal_length)
```

- transform() : 그룹별 계산은 동일하지만 데이터프레임 자체를 변화시킴
- 원본 프레임과 크기는 동일함 
- cut() : 동일 길이로 나누어서 범주를 만들고 그룹에 대한 통계량
- qcut() : 동일 개수로 나누어서 범주를 만들고 그룹에 대한 통계량

```python
# 꽃잎길이가 해당 종 내에서 대/중/소 어느 것에 해당하는지에 대한 프레임 생성
# iris_datasets.groupby(iris_datasets['species'])
iris_datasets.groupby(iris_datasets.species)
```

```python
# type : series
iris_datasets.groupby(iris_datasets.species).petal_length.size()
```

```python
# 시리즈에 대한 데이터를 비교하여 label값 출력
def cat3_petal_length(s) :
    return pd.qcut(s,3,labels=['소','중','대']).astype(str)
```

```python
iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.transform(cat3_petal_length)
display(iris_datasets.head())
iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.apply(cat3_petal_length)
display(iris_datasets.head())
# agg() error
# iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.agg(cat3_petal_length)
# display(iris_datasets.head())
```

```python
# 종별로 꽃잎길이, 꽃잎폭의 평균 구하기 -> 그룹연산 : agg()
iris_datasets.groupby(iris_datasets.species).agg(np.mean).loc[:,['petal_length','petal_width']]
```

```python
# 딕셔너리로 다양한 통계값 출력가능
iris_datasets.groupby(iris_datasets.species).agg({'petal_length' : ['min','max'],
                                                 'petal_width' : 'mean'})
```

```python
# 각각의 시리즈에 대한 min, max 값 출력
iris_datasets.groupby(iris_datasets.species).agg(['min','max'])
```

```python
# 동일한 결과
iris_datasets.groupby(iris_datasets.species).agg([np.min,np.max])
```

- filter() : 그룹 객체에 filter() 적용 시 조건식을 가진 함수 전달 -> 조건이 참인 그룹만 출력
- sql의 where 조건절과 비슷

```python
titanic_datasets = sns.load_dataset('titanic')
titanic_datasets.head()
```

```python
grp_class_titanic = titanic_datasets.groupby(['class'])
grp_class_titanic
```

```python
# 1. 데이터의 개수가 200개 이상인 그룹만 필터링
for key, group in grp_class_titanic :
    print('key : ',key)
    print('number : ',len(group))
    print(group)
    print('\n')
```

```python
def filter_func(x) :
    return len(x) >= 200
```

```python
# 함수 label 이용하여 호출
grp_filter = grp_class_titanic.filter(filter_func) 
grp_filter
```

```python
# 함수 호출이 아닌 람다식 이용
grp_filter = grp_class_titanic.filter(lambda x : len(x) >= 200) 
grp_filter
```

```python
# 2. age 열의 평균이 30보다 작은 그룹만 필터링
# x는 데이터프레임을 의미함
grp_filter = grp_class_titanic.filter(lambda x : x.age.mean()<30)
grp_filter
```

```python
# 반복문 이용하여 출력
# print(grp_filter.index)
for idx in grp_filter.index :
    print(grp_filter.iloc[idx])
```

### pivot(index, columns, value)
### pivot_table(data,values,index,columns,aggfunc='mean')
- 데이터프레임의 pivot_table
- pandas의 pivot_table -> 데이터프레임 정의필수

```python
# 기존 데이터프레임에서 subset
df = titanic_datasets.loc[:,['sex','pclass']]
df
```

```python
df = titanic_datasets[['sex','pclass']]
df
```

```python
# 새로운 데이터프레임 생성
df = pd.DataFrame(titanic_datasets, columns=['sex','pclass'])
df
```

```python
# 성별과 객실등급에 따른 승객 수 집계
# groupby
# 멀티 인덱스를 갖는 시리즈로 출력
# size 대신 value_counts()를 쓸 수 없는 이유 -> value_counts()는 시리즈, 데이터프레임에서만 사용가능
df = titanic_datasets.groupby(['sex','pclass']).size()
display(df)
print(type(df))
```

```python
# reset_index : 시리즈 -> 데이터프레임
# 배열 인덱스 생성
df = titanic_datasets.groupby(['sex','pclass']).size().reset_index(name='cnt')
display(df)
print(type(df))
```

```python
# pivot(index, columns, value)
# 데이터의 가독성 굿
df.pivot('sex','pclass','cnt')
```

```python
# 성별과 생존여부에 따른 승객 수 집계
df = titanic_datasets.groupby(['sex','survived']).size().reset_index(name='cnt')
df.pivot('sex','survived','cnt')
```

```python
# 성별과 객실등급(pclass)에 따른 승객 수 집계
# pivot_table(, aggfunc = mean | sum)
titanic_datasets['cnt'] = 1 # 집계를 위해 컬럼추가
display(titanic_datasets)
display(titanic_datasets.pivot_table(index='sex',columns='pclass',values='cnt',aggfunc=np.sum))
```

```python
# 성별과 생존여부에 따른 승객 수 집계
display(titanic_datasets.pivot_table(index='sex',columns='survived',values='cnt',aggfunc=np.sum))
```

```python
# 합계와 평균을 동시에 보고 싶을 때
display(titanic_datasets.pivot_table(index='sex',columns='pclass',values='cnt',aggfunc=[np.sum,np.mean]))
```

```python
tips_datasets = sns.load_dataset('tips')
tips_datasets
```

```python
tips_datasets.info()
```

```python
# 식사대금 대비 팁의 비율이 어떤 경우에 가장 높은가
tips_datasets['tip_pct'] = tips_datasets['tip'] / tips_datasets['total_bill']
display(tips_datasets)
display(tips_datasets.describe())
```

```python
# display(tips_datasets.groupby('sex').count())
display(tips_datasets.groupby('sex').size()) 
```

```python
# 성별과 흡연자 구분
tips_datasets.groupby(['sex','smoker']).size()
```

```python
pd.pivot_table(data = tips_datasets,
              index = 'sex',
              columns = 'smoker',
               values = 'tip_pct',
              aggfunc = 'count')
```

```python
# margins 사용 -> 카운트에 대한 행과 열의 합 출력
# pandas 사용한 pivot_table -> 데이터프레임 지정해줘야함
pd.pivot_table(data = tips_datasets,
              index = 'sex',
              columns = 'smoker',
               values = 'tip_pct',
              aggfunc = 'count',
              margins = True)
```

```python
# 성별과 흡연여부에 따른 평균 팁의 비율
# groupby
tips_datasets.groupby(['sex','smoker'])[['tip_pct']].mean()
```

```python
pd.pivot_table(data = tips_datasets,
              index = 'sex',
              columns = 'smoker',
               values = 'tip_pct',
              aggfunc = 'mean',
              margins = True)
```
