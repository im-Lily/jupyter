```python
import pandas as pd
import numpy as np
# warning message ignore
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 데이터 불러오기
ratingDF = pd.read_csv('./data/ratings.dat',sep='::',header=None)
display(ratingDF.head()) # 헤더가 없기 때문에 배열 인덱스 자동으로 생성
movieDF = pd.read_csv('./data/movies.dat',sep='::',header=None)
display(movieDF.head())
userDF = pd.read_csv('./data/users.dat',sep='::',header=None)
display(userDF)
```

```python
display(ratingDF.info())
display(ratingDF)
```

```python
# 데이터프레임에 헤더 추가
ratingDF.columns = ['userId','movieId','rating','timestamp']
print(ratingDF.columns)
display(ratingDF)
```

```python
display(movieDF.info)
display(movieDF)
movieDF.columns = ['movieId','title','genress']
print(movieDF.columns)
```

```python
display(userDF.info())
display(userDF)
userDF02 = userDF.loc[:,[0,1,2]]
display(userDF02)
userDF02.columns = ['userId','gender','age']
display(userDF02)
```

```python
display(ratingDF)
display(movieDF)
display(userDF02)
```

```python
# 공통 컬럼을 이용한 데이터 병합
# concat : 데이터프레임에 대한 병합
# merge : 데이터에 대한 병합
# concat_rating_movie = pd.concat([ratingDF,movieDF],axis=1,join='inner')
# display(concat_rating_movie)
```

```python
# merge_rating_movie = pd.merge(ratingDF,movieDF,left_on='movieId',right_on='movieID',how='inner')
# display(merge_rating_movie)
```

```python
merge_rating_movie = pd.merge(ratingDF,movieDF)
# display(merge_rating_movie)
merge_rating_movie_user = pd.merge(merge_rating_movie,userDF02)
display(merge_rating_movie_user)
```

```python
# userId 별 평점 개수 순위 확인
sort_userid_rating = merge_rating_movie_user.groupby('userId').size()
print(type(sort_userid_rating))
# display(sort_userid_rating)
sort_userid_rating.sort_values(ascending=False)
```

```python
# 평점이 200개 이상인 영화들의 성별 평균 평점
mean_rating = merge_rating_movie_user.pivot_table(index='title',
                                   columns='gender',
                                   values = 'rating',
                                   aggfunc = np.mean)
# display(mean_rating)

# 평균값으로 결측값 대치
# mean_rating['F'] = mean_rating['F'].fillna(mean_rating['F']).mean()
# mean_rating['M'] = mean_rating['M'].fillna(mean_rating['M']).mean()
# print(mean_rating.isnull().sum())
# display(mean_rating)

# display(merge_rating_movie_user)
title_rating = merge_rating_movie_user.groupby('title').size()
# display(title_rating>200)

top_rating_title = mean_rating[title_rating>200]
display(top_rating_title)
```

```python
# 여성이 가장 좋아하는 영화 순위
top_rating_title.sort_values(by='F',ascending=False)
```

```python
# 남녀간의 평균 평점 차이
mean_rating['diff'] = mean_rating['F'] - mean_rating['M']
display(mean_rating)
```
