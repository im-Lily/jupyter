- barplot() : 카테고리 값에 따른 숫자값의 평균과 편차를 표시하는 차트
- 평균은 막대의 높이, 편차는 에러바

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()

# 성별에 따른 생존율성별에 따른 생존율
fig = plt.figure(figsize = (15,5))

ax01 = fig.add_subplot(1,3,1)
ax01.set_title('survived - sex')
sns.barplot(x='sex',y='survived',data=titanic_datasets, ax=ax01)

# 등급에 따른 생존율도 확인가능
ax02 = fig.add_subplot(1,3,2)
# hue = 그룹
sns.barplot(x='sex',y='survived',hue='class',data=titanic_datasets, ax=ax02)

# dodge = 
ax03 = fig.add_subplot(1,3,3)
sns.barplot(x='sex',y='survived',hue='class',dodge=False,palette='Set3',data=titanic_datasets, ax=ax03)

plt.show()
plt.close()
```

- countplot() : 카테고리 값들로 데이터의 빈도 표시

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()

# 선실등급의 빈도 확인
# countplot -> x:카테고리, y:의미없음
fig = plt.figure(figsize = (15,5))

ax01 = fig.add_subplot(1,3,1)
ax01.set_title('class count')
sns.countplot(x='class',data=titanic_datasets, ax=ax01)

# 등급에 따른 생존율도 확인가능
ax02 = fig.add_subplot(1,3,2)
# hue = 그룹
sns.countplot(x='class',hue='who',data=titanic_datasets, ax=ax02)

# dodge = 
ax03 = fig.add_subplot(1,3,3)
sns.countplot(x='class',hue='who',dodge=False,palette='Set3',data=titanic_datasets, ax=ax03)

plt.show()
plt.close()
```

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()

fig = plt.figure(figsize = (15,5))

# 요일별 팁의 빈도 시각화
sns.countplot(x='day',hue='smoker',dodge=False,data=tips_datasets)
plt.title('요일별 팁을 준 횟수')

plt.show()
plt.close()
```

- regplot() : 산점도 - 선형회귀분석에 의한 회귀선을 나타내주는 함수
- x,y : 연속(수치형)변수

```python
titanic_datasets.head()
```

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()


fig = plt.figure(figsize = (15,5))

# 나이에 따른 요금 - 산점도와 회귀선 표시
ax01 = fig.add_subplot(1,2,1)
sns.regplot(x='age',y='fare',data=titanic_datasets,ax=ax01)

ax02 = fig.add_subplot(1,2,2)
sns.regplot(x='age',y='fare',data=titanic_datasets,ax=ax02,fit_reg=False)

plt.show()
plt.close()
```

- distplot() : 히스토그램/커널 밀도 그래프

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()

fig = plt.figure(figsize = (15,5))

# 요금에 대한 분포확인가능
area01 = fig.add_subplot(1,3,1)
sns.distplot(titanic_datasets['fare'],ax=area01)

area02 = fig.add_subplot(1,3,2)
sns.distplot(titanic_datasets['fare'],hist=False,ax=area02)

area03 = fig.add_subplot(1,3,3)
sns.distplot(titanic_datasets['fare'],kde=False,ax=area03)

plt.show()
plt.close()
```

- heatmap() : x,y 범주형
- 매트릭스 형태로 데이터 분류

```python
# 피벗테이블로 범주형 변수를 각각 행과 열로 재구분
# 성별에 따른 선실등급
table = titanic_datasets.pivot_table(index=['sex'],columns=['class'],aggfunc='size')
table
```

```python
sns.heatmap(table,cbar=True,annot=True,fmt='d',linewidth=.5)
```

- 범주형 데이터의 분포 stripplot(), swamplot()

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('dark')
plt.figure()


fig = plt.figure(figsize = (15,5))

area01 = fig.add_subplot(1,2,1)
sns.stripplot(x='class',y='age',data=titanic_datasets,ax=area01)

area02 = fig.add_subplot(1,2,2)
sns.swarmplot(x='class',y='age',data=titanic_datasets,ax=area02)

plt.show()
plt.close()
```
