### xpath
- 절대경로를 이용한 노드 탐색
- 노드 선택(nodename, /(root), //, .(node), .., @(속성))
- <div name='test'>
- driverfind_element_by_xpath('//div[@name]')
- * : 매칭되는 모든 요소 의미, @* : 매칭되는 모든 속성의 노드 의미

```python
lis = driver.find_elements_by_css_selector('#emergencyPrd > div > ul > li')
# len(lis)
for li in lis :
    title = li.find_element_by_css_selector('div > a > div.prd_info > p > span.fs_16')
    price = li.find_element_by_css_selector('div > a > div.prd_info > div > span.price_detail > strong')
    print(title.text)
    print(price.text)
#     print(price.get_attribute('innerHTML')) # .text 결과 출력 안될 시 사용
    print('*'*50)
```

```python
lis = driver.find_elements_by_xpath('//*[@id="emergencyPrd"]/div/ul/li[position()<=2]')
len(lis)

for li in lis :
    title = li.find_element_by_xpath('div/a/div[3]/p/span[2]')
    price = li.find_element_by_xpath('div/a/div[3]/div/span/strong')
    print(title.text)
    print(price.text)
```

### page 처리

```python
from time import sleep, time
from random import randint
from IPython.core.display import clear_output
from requests import get
```

```python
start_time = time()
requests = 0
for i in range(5) :
#     print(i)
    requests += 1
    sleep(randint(1,3))
    current_time = time()
    delayed_time = current_time - start_time
    print('requests {} , time {}'.format(requests, delayed_time))
    clear_output(wait=True) # 출력한 곳 삭제 후 업데이트
```
